```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("wine.csv")
df
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:, 1:14]
y = df['Wine']
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=100)
```

```python
#cv is cross validatioin
bakward_feature_selector = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                    k_features=8,
                                                    forward=False,
                                                    floating=False,
                                                    verbose=2,
                                                    scoring= "accuracy",
                                                    cv= 5).fit(X_train, y_train)
```

```python
bakward_feature_selector.k_feature_idx_
```

```python
bakward_feature_selector.k_feature_names_
```

```python
bakward_feature_selector.k_score_
```

```python
bakward_feature_selector = SequentialFeatureSelector(RandomForestClassifier(n_jobs=-1),
                                                    k_features=(1,13),
                                                    forward=False,
                                                    floating=False,
                                                    verbose=2,
                                                    scoring= "accuracy",
                                                    cv= 5).fit(X_train, y_train)
```

```python
bakward_feature_selector.k_score_
```
